```python
from pathlib import Path
import pandas as pd
from pandas_profiling import ProfileReport
import ipyplot
```

```python
train_df = pd.read_csv('../../data/train.csv')
test_df = pd.read_csv('../../data/test.csv')
train_df.shape, test_df.shape
```
